```python
colunas_para_maiusculo = [
    'descricao',
    'objeto',
    'descricao_comp',
    'orgao',
    'Cliente TOTVS Ativo ou Inativo',
    'razao_social',
    'Solução',
    'Ente',
    'regiao',
    'CNPJ'
]

for col in colunas_para_maiusculo:
    if col in df_totvs.columns and df_totvs[col].dtype == 'object':
        df_totvs[col] = df_totvs[col].apply(lambda x: x.upper() if isinstance(x, str) else x)
    else:
        print(f"Coluna '{col}' não encontrada no DataFrame.")
```

```python
def tipo_de_compra(texto):
    texto = str(texto).upper()
    if 'SERVIÇO' in texto or 'SERVICO' in texto:
        return 'SERVIÇO'
    elif 'SOFTWARE' in texto:
        return 'SOFTWARE'
    else:
        return 'OUTROS'

def tipo_de_software(texto):
    texto = str(texto).upper()
    if 'ERP' in texto:
        return 'ERP'
    elif 'SOFTWARE' in texto:
        return 'OUTROS'
    else:
        return 'OUTROS'

def modalidades(texto):
    texto = str(texto).upper()
    if 'CLOUD' in texto or 'NUVEM' in texto:
        return 'NUVEM/CLOUD'
    elif 'ON PREMISE' in texto or 'ON-PREMISE' in texto:
        return 'ON PREMISE'
    else:
        return 'OUTROS'

def tipo_de_nuvem(texto):
    texto = str(texto).upper()
    if 'IAAS' in texto:
        return 'IAAS'
    elif 'PAAS' in texto:
        return 'PAAS'
    elif 'SAAS' in texto:
        return 'SAAS'
    else:
        return 'OUTROS'

def classificacao_tipo_erp(texto):
    texto = str(texto).lower()
    if "educa" in texto:
        return "Educacional"
    elif "saude" in texto:
        return "Saúde"
    elif "financeiro" in texto or "contábil" in texto or "contabil" in texto:
        return "Financeiro"
    elif "gestão" in texto or "gestao" in texto:
        return "Gestão"
    elif "folha" in texto or "rh" in texto:
        return "RH"
    elif "ponto" in texto:
        return "Controle de Ponto"
    elif "jurídico" in texto or "juridico" in texto:
        return "Jurídico"
    elif "hotelaria" in texto:
        return "Hotelaria"
    elif "consulting" in texto or "consultoria" in texto:
        return "Consultoria"
    elif "rd" in texto:
        return "Relacionamento Digital"
    elif "ia" in texto:
        return "Inteligência Artificial"
    elif "desenvolvimento" in texto or "desen" in texto:
        return "Desenvolvimento"
    else:
        return "OUTROS"


df_totvs['Tipo de Compra'] = df_totvs['objeto'].apply(tipo_de_compra)
df_totvs['Classificação de Software'] = df_totvs['objeto'].apply(tipo_de_software)
df_totvs['Modalidades'] = df_totvs['objeto'].apply(modalidades)
df_totvs['Classificação Cloud'] = df_totvs['objeto'].apply(tipo_de_nuvem)
df_totvs['Classificação Tipo de ERP'] = df_totvs['Solução'].apply(classificacao_tipo_erp)

df_totvs = df_totvs.loc[:, ~df_totvs.columns.duplicated()]

colunas = df_totvs.columns.tolist()
colunas_classificadas = ['Tipo de Compra', 'Classificação de Software', 'Modalidades', 'Classificação Cloud', 'Classificação Tipo de ERP']

for nova_coluna in colunas_classificadas:
    if nova_coluna in colunas:
        colunas.remove(nova_coluna)

pos_objeto = colunas.index('objeto')
colunas[pos_objeto + 1:pos_objeto + 1] = colunas_classificadas
df_totvs = df_totvs[colunas]

resumo_classificacoes = pd.DataFrame()
total_linhas = len(df_totvs)

for coluna in colunas_classificadas:
    contagem = df_totvs[coluna].value_counts(dropna=False).reset_index()
    contagem.columns = ['Categoria', 'Frequência']
    contagem['%'] = (contagem['Frequência'] / total_linhas * 100).round(2)
    contagem['Coluna'] = coluna
    resumo_classificacoes = pd.concat([resumo_classificacoes, contagem], ignore_index=True)

resumo_classificacoes = resumo_classificacoes[['Coluna', 'Categoria', 'Frequência', '%']]

display(resumo_classificacoes)
```

```python
import pandas as pd
import numpy as np
df_totvs.replace(r'^\s*$', np.nan, regex=True, inplace=True)
total_registros = df_totvs.shape[0]
total_colunas = df_totvs.shape[1]
coluna_data = 'data'
if coluna_data in df_totvs.columns:
    df_totvs[coluna_data] = df_totvs[coluna_data].replace(r'^\s*$', np.nan, regex=True)

    df_totvs[coluna_data] = pd.to_numeric(df_totvs[coluna_data], errors='coerce')

    if df_totvs[coluna_data].notnull().any():
        dado_mais_antigo = int(df_totvs[coluna_data].min())
        dado_mais_recente = int(df_totvs[coluna_data].max())
    else:
        dado_mais_antigo = 'Dados inválidos'
        dado_mais_recente = 'Dados inválidos'
else:
    dado_mais_antigo = 'N/A'
    dado_mais_recente = 'N/A'

missing_percent = df_totvs.isnull().mean()

colunas_mais_50_missing = (missing_percent > 0.5).sum()
colunas_com_nulos_ate_50 = ((missing_percent > 0) & (missing_percent <= 0.5)).sum()

colunas_id = [
    col for col in df_totvs.columns
    if 'id' in col.upper() or 'c1' in col.upper() or 'c2' in col.upper()
]

print("Resumo:")
print(f"Total de registros: {total_registros}")
print(f"Total de colunas: {total_colunas}")
print(f"Dado mais antigo: {dado_mais_antigo}")
print(f"Dado mais recente: {dado_mais_recente}")
print(f"Colunas com > 50% missing: {colunas_mais_50_missing}")
print(f"Colunas com <= 50% missing: {colunas_com_nulos_ate_50}")
if colunas_id:
    print(f"Colunas identificadas como IDs: {colunas_id}")
missing_data = df_totvs.isnull().sum().to_frame(name='N')
missing_data['%'] = (missing_data['N'] / total_registros * 100).round(2)
missing_data = missing_data.reset_index().rename(columns={'index': 'Coluna'})
missing_data = missing_data[missing_data['N'] > 0]
missing_data = missing_data.sort_values(by='N', ascending=False)

if not missing_data.empty:
    print("\n Colunas com valores nulos (incluindo campos vazios ou espaços):")
    print(missing_data.to_string(index=False))
else:
    print("\n Nenhuma coluna com valores nulos ou vazios.")
```

```python
import pandas as pd
import numpy as np
from IPython.display import display

valor_numerico = (
    df_totvs['valor_total']
    .astype(str)
    .str.replace("R\$\s?", "", regex=True)
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
    .astype(float)
)

resumo = pd.DataFrame({
    'Estatística': ['Média', 'Máximo', 'Mínimo', 'Mediana', 'Soma', 'Desvio Padrão'],
    'Valor': [
        valor_numerico.mean(),
        valor_numerico.max(),
        valor_numerico.min(),
        valor_numerico.median(),
        valor_numerico.sum(),
        valor_numerico.std()
    ]
})

resumo['Valor'] = resumo['Valor'].apply(
    lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
)

estatisticas_avancadas = valor_numerico.describe(percentiles=[.25, .5, .75])

estatisticas_formatadas = pd.DataFrame({
    'Estatística': estatisticas_avancadas.index,
    'Valor': estatisticas_avancadas.apply(
        lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
    ).values
})

print("\n Resumo Estatístico:")
display(resumo)

print("\n Estatísticas Descritivas:")
display(estatisticas_formatadas)
```

```python
!pip install openpyxl

import pandas as pd
import re

caminho_arquivo = '/content/Cópia de Estudo de Mercado V.05.03.2025 (cnpj).xlsx'

xls = pd.ExcelFile(caminho_arquivo)
df_totvs = pd.read_excel(xls, sheet_name=0)
df_fonte_precos = pd.read_excel(xls, sheet_name="CNPJs Fonte de Preços")
df_clientes_totvs = pd.read_excel(xls, sheet_name="Clientes TOTVS CNPJ")

def validar_cnpj(cnpj):
    cnpj = re.sub(r'\D', '', str(cnpj))
    if len(cnpj) != 14 or len(set(cnpj)) == 1:
        return False
    def calcular_digito(cnpj_base, pesos):
        soma = sum(int(d) * p for d, p in zip(cnpj_base, pesos))
        resto = soma % 11
        return '0' if resto < 2 else str(11 - resto)
    cnpj_base = cnpj[:12]
    digito1 = calcular_digito(cnpj_base, [5, 4, 3, 2, 9, 8, 7, 6, 5, 4, 3, 2])
    cnpj_base += digito1
    digito2 = calcular_digito(cnpj_base, [6, 5, 4, 3, 2, 9, 8, 7, 6, 5, 4, 3, 2])
    return cnpj[-2:] == digito1 + digito2

def formatar_cnpj(cnpj):
    cnpj = re.sub(r'\D', '', str(cnpj)).zfill(14)
    return f"{cnpj[:2]}.{cnpj[2:5]}.{cnpj[5:8]}/{cnpj[8:12]}-{cnpj[12:14]}"

coluna_cnpj_totvs = next((col for col in df_totvs.columns if "cnpj" in col.lower()), None)
coluna_cnpj_fonte = next((col for col in df_fonte_precos.columns if "cnpj" in col.lower()), None)
coluna_cnpj_clientes = next((col for col in df_clientes_totvs.columns if "cnpj" in col.lower()), None)

if coluna_cnpj_totvs and coluna_cnpj_fonte and coluna_cnpj_clientes:
    for df, coluna in [(df_totvs, coluna_cnpj_totvs), (df_fonte_precos, coluna_cnpj_fonte), (df_clientes_totvs, coluna_cnpj_clientes)]:
        df[coluna] = df[coluna].astype(str).str.strip().apply(lambda x: re.sub(r'\D', '', x).zfill(14))
        df["CNPJ_formatado"] = df[coluna].apply(formatar_cnpj)

    df_totvs["CNPJ_valido"] = df_totvs[coluna_cnpj_totvs].apply(validar_cnpj)
    nao_localizados = df_totvs[coluna_cnpj_totvs].isin(["", "NÃO LOCALIZADO", "nan"]) | df_totvs[coluna_cnpj_totvs].isnull()
    df_totvs.loc[nao_localizados, "CNPJ_valido"] = False

    total = len(df_totvs)
    unicos = df_totvs[coluna_cnpj_totvs].nunique()
    duplicados = total - unicos
    total_invalidos = df_totvs[~df_totvs["CNPJ_valido"]].shape[0]
    cnpjs_invalidos = df_totvs[~df_totvs["CNPJ_valido"] & ~nao_localizados][coluna_cnpj_totvs].value_counts()
    total_nao_localizados = nao_localizados.sum()
    cnpj_counts = df_totvs[coluna_cnpj_totvs].value_counts()
    cnpjs_duplicados = cnpj_counts[cnpj_counts > 1].index
    df_duplicados = df_totvs[df_totvs[coluna_cnpj_totvs].isin(cnpjs_duplicados)]
    duplicados_validos = df_duplicados[df_duplicados["CNPJ_valido"]].shape[0]
    duplicados_invalidos = df_duplicados[~df_duplicados["CNPJ_valido"]].shape[0]

    resultado = {
        "Total de Registros": total,
        "CNPJs Únicos": unicos,
        "CNPJs Duplicados Total": duplicados,
        "Total de Registros com CNPJ Inválido": total_invalidos,
        "CNPJs Inválidos Únicos": len(cnpjs_invalidos),
        "Top CNPJs Inválidos": cnpjs_invalidos.head(10).to_dict(),
        "Total de CNPJs Não Localizados": total_nao_localizados,
        "CNPJs Duplicados e Válidos": duplicados_validos,
        "CNPJs Duplicados e Inválidos": duplicados_invalidos
    }

    for chave, valor in resultado.items():
        print(f"{chave}: {valor}")
    df_cruzado_fonte = pd.merge(df_totvs, df_fonte_precos, on="CNPJ_formatado", how="inner")
    df_cruzado_clientes = pd.merge(df_totvs, df_clientes_totvs, on="CNPJ_formatado", how="inner")
else:
    print("Coluna de CNPJ não encontrada em uma das abas.")
```
